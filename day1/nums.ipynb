```python
import re
```

```python
# read in input.txt
lines = [line.strip() for line in open('input.txt', 'r')]
```

```python
totalVal = []
```

```python
for line in lines:
    digits = re.sub(r'\D', '', line)
    if len(digits) > 1:
        digits = [int(d) for d in digits]
        total = str(digits[0]) + str(digits[-1:][0])
        totalVal.append(int(total))
    #break
    elif len(digits) == 1:
        total = str(digits[0]) + str(digits[0])
        totalVal.append(int(total))
```

```python
sum(totalVal)
```

```python
totalVal2 = []
```

```python
word2num = {'one':'1', 'two':'2', 'three':3, 'four':4, 'five':5, 'six':6, 
            'seven':7, 'eight':8, 'nine':9}
```

```python
for line in lines:
    # replace string with val in word2num
    for word in word2num.keys():
        line = line.replace(word, f'{word[0]}{str(word2num[word])}{word[-1]}')
    digits = re.sub(r'\D', '', line)

    if len(digits) > 1:
        digits = [int(d) for d in digits]
        total = str(digits[0]) + str(digits[-1:][0])
        print(total)
        totalVal2.append(int(total))
    elif len(digits) == 1:
        total = str(digits[0]) + str(digits[0])
        print(total)
        totalVal2.append(int(total))
```
